**General integration (quad):** The function quad is provided to integrate a function of one variable between two points. The first argument to quad is a â€œcallableâ€ Python object (i.e., a function, method, or class instance). The next two arguments are the limits of integration. 

The return value is a tuple, with the first element holding the estimated value of the integral and the second element holding an upper bound on the error.

**General multiple integration (dblquad, tplquad, nquad):**
The mechanics for double and triple integration have been wrapped up into the functions dblquad and tplquad. These functions take the function to integrate and four, or six arguments, respectively. The limits of all inner integrals need to be defined as functions.

```python
from scipy.integrate import quad

def f(D):
    return 20 * np.exp(-20*(D-12.5))

# P(D > 12.6)           find the proportion. 
proportion  =  round(quad(f, 12.6 , np.inf)[0],3)
proportion
```

Problem Statement 5:

A company manufactures LED bulbs with a faulty rate of 30%. If I randomly select 6
chosen LEDs, what is the probability of having 2 faulty LEDs in my sample?
Calculate the average value of this process. Also evaluate the standard deviation
associated with it.


```python
#  nCr 
def nCr(n, r):   
    return (  fact(n) / (fact(r) * fact(n - r)) ) 
  
# Factorial of n 
def fact(n): 
    res = 1    
    for i in range(2, n+1):
        res = res * i           
    return res 

# SOLUTION:

# This is a binomial distribution because there are only 2 possible outcomes - Faulty-LED or Good-LED.


P_FaultyLED = 0.3  
P_GoodLED  = 0.7

#  sample
n = 6

P_FaultyLED_is_2 = nCr(6, 2) * (P_FaultyLED**2) *  (P_GoodLED**4)

print("Probability of having 2 faulty LEDs:", round(P_FaultyLED_is_2,3))
```

Problem Statement 6:

Gaurav and Barakha are both preparing for entrance exams. Gaurav attempts to
solve 8 questions per day with a correction rate of 75%, while Barakha averages
around 12 questions per day with a correction rate of 45%. 
- What is the probability that each of them will solve 5 questions correctly? 
- What happens in cases of 4 and 6 correct solutions? 
- What do you infer from it? 
- What are the two main governing factors affecting their ability to solve questions correctly? 
- Give a pictorial representation of the same to validate your answer.

```python
# SOLUTION:

# This is a binomial distribution because there are only 2 possible outcomes - Correct Answer / Wrong Anaswer.

def CorrectedProbability(n_sample,correct_rate,req_correct):
    pr = nCr(n_sample, req_correct) * (correct_rate**req_correct) *  ((1-correct_rate)**(n_sample-req_correct))
    return round(pr,3)


print("Probability of Gaurav with 5 Questions correct:", CorrectedProbability(8,0.75,5))
print("Probability of Baraka with 5 Questions correct:", CorrectedProbability(12,0.45,5))
```

```python
print("Probability of Gaurav with 4 Questions correct:", CorrectedProbability(8,0.75,4))
print("Probability of Baraka with 4 Questions correct:", CorrectedProbability(12,0.45,4))
```

```python
print("Probability of Gaurav with 6 Questions correct:", CorrectedProbability(8,0.75,6))
print("Probability of Baraka with 6 Questions correct:", CorrectedProbability(12,0.45,6))
```
